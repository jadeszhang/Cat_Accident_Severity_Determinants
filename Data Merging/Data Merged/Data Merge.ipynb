```python
import pandas as pd
```

```python
accident = pd.read_csv('ACCIDENT.csv')
damage = pd.read_csv('DAMAGE.csv') ##
acc_aux = pd.read_csv('ACC_AUX.csv')
cevent = pd.read_csv('CEVENT.csv')
maneuver = pd.read_csv('MANEUVER.csv') ##
factor = pd.read_csv('FACTOR.csv') ##
distract = pd.read_csv('DISTRACT.csv') ##
drimpair = pd.read_csv('DRIMPAIR.csv') ##
parkwork = pd.read_csv ('PARKWORK.csv') ##
pbtype = pd.read_csv ('PBTYPE.csv')
nmprior = pd.read_csv('NMPRIOR.csv') ##
nmimpair = pd.read_csv('NMIMPAIR.csv') ##
nmcrash = pd.read_csv('NMCRASH.csv') ##
per_aux = pd.read_csv('PER_AUX.csv')
person = pd.read_csv('PERSON.csv') ##
veh_aux = pd.read_csv('VEH_AUX.csv')
vevent = pd.read_csv('VEVENT.csv')##
vsoe = pd.read_csv('VSOE.csv')##
violatn = pd.read_csv('VIOLATN.csv') ##
vision = pd.read_csv('VISION.csv') ##
vehicle = pd.read_csv('VEHICLE.csv') ##
safetyeq = pd.read_csv('SAFETYEQ.csv') ##
```

first merge vehicle info

```python
# vehicle level

# vehicle: This data file contains information describing the in-transport
# motor vehicles and the drivers of in-transport motor vehicle who are involved in the
# crash. There is one record per in-transport motor vehicle. Parked and working vehicle
# information is in the Parkwork data file

# parkwork: This data file contains information about parked and working
# vehicles that were involved in FARS crashes

# vindecode: This data file contains vehicle descriptors for all vehicles,
# mainly passenger vehicles, trucks and motorcycles, based on the vehicleâ€™s VIN which is
# decoded using the VINtelligence program. There is one record per vehicle.
```

```python
# merge vehicle and parkwork by casenum
vehicle_info = pd.concat([vehicle,parkwork], axis = 0,sort = False, ignore_index = True)

# merge with damage 
# damage: This data file contains information about all of the areas on
# this vehicle that were damaged in the crash. There is one record per damaged area.
damage['MDAREAS'] = damage['MDAREAS'].astype(str)
damage = damage.groupby(['CASENUM','VEH_NO'])['MDAREAS'].apply(','.join).reset_index()
vehicle_info = vehicle_info.merge(damage, how = 'inner', on = ['CASENUM','VEH_NO'])

# merge with factor
# factor: This data filecontains information about vehiclecircumstanceswhich may have contributed to the crash.  There is at least one record per in-transport motor vehicle. Each factor is a separate record.  
# merge MFACTOR into the current dataset by veh_no & casenum
vehicle_info = vehicle_info.merge(factor[['CASENUM','VEH_NO','MFACTOR']], how = 'outer', on = ['CASENUM','VEH_NO'])

# merge with distract
# distract: This data file contains information about driver distractions.  There is at least one record per in-transport motor vehicle. Each distraction is a separate record. 
# merge MDRDSTRD into the current dataset by veh_no & casenum
vehicle_info = vehicle_info.merge(distract[['CASENUM','VEH_NO','MDRDSTRD']], how = 'outer', on = ['CASENUM','VEH_NO'])

# merge with maneuver
#  This data file contains information about actions taken by
# the driver to avoid something or someone in the road. There is at least one record per
# in-transport motor vehicle. Each maneuver is a separate record. 
# merge MDRMANAV into the current dataset by veh_no & casenum
vehicle_info = vehicle_info.merge(maneuver[['CASENUM','VEH_NO','MDRMANAV']], how = 'outer', on = ['CASENUM','VEH_NO'])

# merge with drimpair
vehicle_info = vehicle_info.merge(drimpair[['CASENUM','VEH_NO','DRIMPAIR']], how = 'outer', on = ['CASENUM','VEH_NO'])

# merge with violation
vehicle_info = vehicle_info.merge(violatn[['CASENUM','VEH_NO','MVIOLATN']], how = 'outer', on = ['CASENUM','VEH_NO'])

# merge with vision 
vehicle_info = vehicle_info.merge(vision[['CASENUM','VEH_NO','MVISOBSC']], how = 'outer', on = ['CASENUM','VEH_NO'])



vehicle_info = vehicle_info.sort_values(by = ['CASENUM'])
```

```python
vehicle_info.to_csv('vehicle_info.csv')
```

merge person info

```python
# person: This data file contains information describing all persons involved in the crash including motorists 
# pbtype: This data file contains information about crashes between
# motor vehicles and pedestrians, people on personal conveyances and bicyclists
# safetype: This data file contains information about safety equipment
# used by people who are not occupants of motor vehicles. T

# merge person and pbtype
person_info = pd.concat([person, pbtype], axis = 0,sort = False, ignore_index = True)

# merge safetype
person_info = person_info.merge(safetyeq[['CASENUM','VEH_NO','PER_NO', 'NMHELMET',
                                         'NMPROPAD','NMOTHPRO','NMREFCLO','NMLIGHT','NMOTHPRE']], how = 'outer', on = ['CASENUM','VEH_NO','PER_NO'])

# merge drugs???

# merge nmcrash
person_info = person_info.merge(nmcrash[['CASENUM','VEH_NO','PER_NO', 'MTM_CRSH']], how = 'outer', on = ['CASENUM','VEH_NO','PER_NO'])

# merge nmimpair
person_info = person_info.merge(nmimpair[['CASENUM','VEH_NO','PER_NO','NMIMPAIR']], how = 'outer', on = ['CASENUM','VEH_NO','PER_NO'])

# merge nmprior
person_info = person_info.merge(nmprior[['CASENUM','VEH_NO','PER_NO','MPR_ACT']], how = 'outer', on = ['CASENUM','VEH_NO','PER_NO'])


person_info = person_info.sort_values(by = ['CASENUM'])

```

```python
person_info.tail(50)
person_info.to_csv('person_info.csv')
```

merge crash & event info


```python
# cevent: This data file contains information for all of the qualifying
# events (i.e., both harmful and non-harmful involving in-transport motor vehicles) which
# occurred in the crash

# vevent:This data file contains the sequence of events for each intransport motor vehicle involved in the crash. 

# vsoe: This data file contains the sequence of events for each intransport motor vehicle involved in the crash. 

# merge event and accident
accident_info = accident.merge(vevent, how = 'outer', on = 'CASENUM',suffixes=('', '_y'))
accident_info = accident_info.drop(accident_info.filter(regex='_y$').columns.tolist(),axis=1)

accident_info.head()
```

```python
accident_info.to_csv('accident_info.csv')
```

```python
# merge three data files
full_data = accident_info.merge(person_info, how = 'outer', on = ['CASENUM','VEH_NO'], suffixes=('', '_y'))
full_data = full_data.drop(full_data.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)

full_data = full_data.merge(vehicle_info, how = 'outer', on = ['CASENUM','VEH_NO'],suffixes=('', '_y'))
full_data = full_data.drop(full_data.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)

full_data.to_csv('full_data.csv')
```
