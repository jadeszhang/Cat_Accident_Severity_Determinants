# 1. Accident Related Data Import
## Accident Related Data will be sperated into three level(case level, vehicle level, person level)
## In this project, the data will be merged in vehicle level, which means per record including one vehicle

```python
### Accident level data
ACCIDENT = pd.read_csv('ACCIDENT.csv')

### Vehcicle level data
VEVENT = pd.read_csv('VEVENT.csv')
DAMAGE = pd.read_csv('DAMAGE.csv')
DISTRACT = pd.read_csv('DISTRACT.csv')
# vehicle description 
VINDECODE = pd.read_csv('VINDECODE.csv')
#vehicle circumstances which may have contributed to the crash
FACTOR = pd.read_csv('FACTOR.csv')
#action taken by the driver to avoid something or someone
MANEUVER = pd.read_csv('MANEUVER.csv')



### Person level data
PERSON = pd.read_csv('PERSON.csv')
# phicisal impairments of  driver 
DRIMPAIR = pd.read_csv('DRIMPAIR.csv')




```

# 1.1 Data Merge

```python
### merge data on vehicle level
vehicle_merged = pd.merge(VEVENT, DAMAGE, on=["STATE","ST_CASE",'VEH_NO'], how= 'left')
vehicle_merged = pd.merge(vehicle_merged, DISTRACT, on=["STATE","ST_CASE",'VEH_NO'], how= 'left')
vehicle_merged = pd.merge(vehicle_merged, VINDECODE, on=["STATE","ST_CASE",'VEH_NO'], how= 'left')
vehicle_merged = pd.merge(vehicle_merged, FACTOR, on=["STATE","ST_CASE",'VEH_NO'], how= 'left')
vehicle_merged = pd.merge(vehicle_merged, MANEUVER, on=["STATE","ST_CASE",'VEH_NO'], how= 'left')
vehicle_merged = pd.merge(vehicle_merged, DRIMPAIR, on=["STATE","ST_CASE",'VEH_NO'], how= 'left')
```

```python
vehicle_merged
```

```python
# Create FIPS Code to help merge demographic data later
ACCIDENT['FIPStxt'] = ACCIDENT["STATE"].astype(str).str.zfill(2) + ACCIDENT["COUNTY"].astype(str).str.zfill(3)
```

```python
ACCIDENT
```

```python
# accident and vehicle
accident_vehicle_merged = pd.merge(vehicle_merged, ACCIDENT, on=["STATE","ST_CASE"], how= 'left')
```

```python
accident_vehicle_merged
```

# 2. Import demographic data

```python
# demographic data including:
# 2. Population based on county
# 3. Poverty based on county
# 4. Unemployment based on county
# 5. Education based on county

population = pd.read_excel('PopulationEstimates.xlsx')
poverty = pd.read_excel('PovertyEstimates.xls')
unemployment = pd.read_excel('Unemployment.xls')
education = pd.read_excel('Education.xls')
```

## 2.1 Population data clean

```python
# drop the useless row and reset the index for population
population.columns = population.iloc[1]
population = population.drop([0,1]).reset_index(drop=True)
population
```

```python
# Extracting Missing Count and Unique Count by Column
unique_count = []
for x in population.columns:
    unique_count.append([x,len(population[x].unique()),population[x].isnull().sum()])

pd.DataFrame(unique_count, columns=["Column","Unique","Missing"]).set_index("Column").T
```

## 2.2 Poverty data clean

```python
# drop the useless rows and reset the index for poverty
poverty.columns = poverty.iloc[3]
poverty = poverty.drop([0,2]).reset_index(drop=True)
poverty = poverty.drop([0]).reset_index(drop=True)
poverty = poverty.drop([0]).reset_index(drop=True)
poverty
```

```python
# Extracting Missing Count and Unique Count by Column
unique_count = []
for x in poverty.columns:
    unique_count.append([x,len(poverty[x].unique()),poverty[x].isnull().sum()])

pd.DataFrame(unique_count, columns=["Column","Unique","Missing"]).set_index("Column").T
```

```python
# drop the useless row and reset the index for unemployment
unemployment = unemployment.drop([0,2]).reset_index(drop=True)
unemployment.columns = unemployment.iloc[4]
unemployment = unemployment.drop([0]).reset_index(drop=True)
unemployment = unemployment.drop([0,1]).reset_index(drop=True)
unemployment = unemployment.drop([0]).reset_index(drop=True)
unemployment = unemployment.drop([0]).reset_index(drop=True)
unemployment
```

## 2.5 Select useful columns
### in this stage, only the FIPS code and data in 2018 will be kept in each demographic dataframe

```python
# parameters: original df(dataframe), keyword(string)
def column_selection (df, keyword):
    # FIPS_code
    FIPS_code_cols = ['FIPStxt']
    # feature
    keyword_cols = [x for x in df.columns[df.columns.str.contains(keyword)]]
    # attach 
    cols = FIPS_code_cols + keyword_cols
    # the new dataframe
    dataframe = df[cols]
    
    return dataframe
```

```python
# use column select function to filter useful columns
education_2018 = column_selection(education,'18')
unemployment_2018 = column_selection(unemployment,'18')
population_2018 = column_selection(population,'18')
poverty_2018 = column_selection(poverty,'18')
```

```python
population_2018
```

# 3. Merge accident data and demographic data

```python
# Merge the data based on the FIPS code
all_merged = pd.merge(accident_vehicle_merged, education_2018, on = ['FIPStxt'], how= 'left')
all_merged = pd.merge(all_merged, unemployment_2018, on = ['FIPStxt'], how= 'left')
all_merged = pd.merge(all_merged, population_2018, on = ['FIPStxt'], how= 'left')
all_merged = pd.merge(all_merged, poverty_2018, on = ['FIPStxt'], how= 'left')
```

```python
all_merged
```

```python
# Extracting Missing Count and Unique Count by Column
unique_count = []
for x in all_merged.columns:
    unique_count.append([x,len(all_merged[x].unique()),all_merged[x].isnull().sum()])

pd.DataFrame(unique_count, columns=["Column","Unique","Missing"]).set_index("Column").T
```
